```python
class Node:
    def __init__(self, freq, symbol, left=None, right=None):
        # frequency of symbol
        self.freq = freq

        # symbol
        self.symbol = symbol

        # left node
        self.left = left

        # right node
        self.right = right

        # Code of the Node. tree direction 0->1
        self.code = ''
    def __str__(self):
        return '<' + str(self.symbol)+',' + str(self.freq) + '>'

    def print_tree(self):
        s='('+str(self)+' '
        if self.left!=None:
            s+=self.left.print_tree()
        if self.right!=None:
            s+=self.right.print_tree()
        s+=' )'
        return s


class Huffman:



  #calc_freq
  def calc_freq(self, texto):

    lista_elementos = dict();
    for char in texto:
      if char in lista_elementos:
        lista_elementos[char] = lista_elementos[char] + 1
      else:
        lista_elementos[char] = 1

    return(lista_elementos)


  #codes_to_tree
  def codes_to_tree(self, codes):
    tree = Node(0, '')

    for char in codes.keys():
      actual_node = tree;
      item_count = 0
      for item in codes[char]:
        item_count = item_count + 1
        # go left
        if item == '0':
          if actual_node.left != None:
            #print('indo para esq')
            actual_node = actual_node.left
          else:
            if item_count == len(codes[char]):
              #print('adding node to ' + char)
              actual_node.left = Node(0, char)
            else:
              #print('add novo filho esq')
              actual_node.left = Node(0, '')
            actual_node = actual_node.left
        # go right
        else:
          if actual_node.right != None:
            #print('indo para dir')
            actual_node = actual_node.right
          else:
            if item_count == len(codes[char]):
              actual_node.right = Node(0, char)
              #print('adding node to ' + char)
            else:
              #print('add novo filho dir')
              actual_node.right = Node(0, '')
            actual_node = actual_node.right

    return tree



  # calc_codes
  def calc_codes(self, node,caminho='', codes=dict()):
      if node.left!=None:
          self.calc_codes(node.left,caminho+'0', codes)
      if node.right!=None:
          self.calc_codes(node.right,caminho+'1', codes)
      if node.left==None and node.right==None:
          codes[node.symbol]=caminho
      return codes



  #debug_node_list
  def debug_node_list(self, nodes):
    out = 'debug nodes: '
    for item in nodes:
      out = out + '<' + str(item.freq) + ',' + item.symbol + '> '
    print(out)


  def build_tree_from_frequency(self, freq_list):

    #ordenando lista de frequencia do char menos usado para o mais usado
    sorted_freq_list = dict(sorted(freq_list.items(), key=lambda item: item[1]))
    sorted_freq_list
    #print(sorted_freq_list)

    #criando lista de nodes
    nodearray = []
    newNode = None
    for item in sorted_freq_list.keys():
      nodearray.append(Node(sorted_freq_list[item], item))

    nodearray.reverse()
    #debug_node_list(nodearray)


    #print('criando arvore')

    while (len(nodearray) > 1):

      node1 = nodearray.pop()
      node2 = nodearray.pop()

      newNode = Node(node1.freq + node2.freq, '')

      if node1.freq > node2.freq:
        newNode.left = node1
        newNode.right = node2
      else:
        newNode.left = node2
        newNode.right = node1

      #print('frequencia do novo node: ' + str(newNode.freq))

      #adicionando o novo no no array de nodes, na posicao ordenada de frequencia
      i = len(nodearray);
      while i > 0:
        i = i-1
        #print(nodearray[i].freq)

        #procura local para inserir no meio do array
        if newNode.freq <= nodearray[i].freq:
          #print('achei local para inserir em ' + str(i))
          nodearray = nodearray[0:i+1] + [newNode] + nodearray[i+1:]
          break

        #se chegou no fim, eh o maior numero, insere aqui mesmo
        elif i == 0:
          nodearray = [newNode] + nodearray[0:]
          break

      #debug_node_list(nodearray)

    #print('construcao finalizada')
    return newNode



  def encode (self, texto):

    print('entrada:           "' + texto + '"')

    #calcula frequencia de cada char
    lista_frequencia = self.calc_freq(texto)
    print('frequencias:       ' + str(lista_frequencia))

    #constroi arvore huffman
    t = self.build_tree_from_frequency(lista_frequencia)
    print('arvore:            ' + t.print_tree())

    #constroi dicionario
    codes = self.calc_codes(t)
    print('dicionario:        ' + str(codes))

    #monta nova lista de bits
    bits_output = ''
    for char in texto:
      bits_output = bits_output + codes[char]
    bits_output_size = len(bits_output)
    print('bits_output:       ' + str(bits_output))
    print('bits_output_size:  ' + str(bits_output_size))

    ##todo: salvar em arquivo de saida


#
#TESTES
#
#encode_huffman('gustavo esta testando o algoritmo de huffman')
huff = Huffman()

huff.encode('AAAAAAABCCCCCCDDEEEEE')


```
